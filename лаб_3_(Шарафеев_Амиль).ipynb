```python
from random import randint
N = int(input('Ввести N: '))
mas = []
for i in range (0, N):
  mas.append(randint(0, 10000))
count = 0
flag = False
for i in range(1, N):
    if mas[i] > mas[i - 1]:
        if not flag:
            count += 1
            flag = True
    else:
        flag = False
print(mas)
print(count)
```

```python
def word_frequency(text):
    words = text.split()
    frequency = {}
    for word in words:
        word = word.lower().strip('.,!?()[]{}"\'')
        if word:
            frequency[word] = frequency.get(word, 0) + 1
    return frequency

# Пример использования:
text = "Привет, как дела? Привет!"
print(word_frequency(text))
```

```python
def dict_reverse(dict):
  reverted = {}
  for key, value in dict.items():
    if value in reverted:
      reverted[value].append(key)
    else:
      reverted[value] = [key]
  return reverted

a = {'a': 1, 'b': 2, 'c': 3}
b = dict_reverse(a)
print(a)
print(b)
```

```python
def dict_merge(dict1, dict2):
  merged = dict1.copy()
  for key, value in dict2.items():
      if key in merged:
          merged[key] += value
      else:
          merged[key] = value
  return merged

users = {"+1111111": "Annya", "+3333333": "Zukhra"}
users2 = {"+2222222": "Masha", "+3333333": "Dasha"}
users = dict_merge(users, users2)
print(users)
```

```python
def dict_porog(dict, porog):
  newdict = {}
  for key, value in dict.items():
    if value > porog:
      newdict[key] = value
  return newdict

a = {'a': 1, 'b': 2, 'c': 3, 'd': 4, 'e': 5, 'f': 6}
b = dict_porog(a, 3)
print(a)
print(b)
```

```python
def dict_max(dict):
  max_key = None
  max_value = None
  for key, value in dict.items():
      if max_value is None or value > max_value:
          max_value = value
          max_key = key
  return max_key, max_value

a = {'a': 1, 'b': 2, 'c': 3, 'd': 4, 'e': 5, 'f': 6}
b, c = dict_max(a)
print(b,c)
```

```python
def length_grouping(dict):
  result = {}
  for key in dict:
      length = len(key)
      if length in result:
          result[length].append(key)
      else:
          result[length] = [key]
  return result

a = {'1': 'Max', '22': 'Alex', '4444': 'Richard', '55555': 'Hans', '66666':'Magnus', '77777':'Chet', '333': 'Jack'}
b = length_grouping(a)
print(b)
```

```python
def merge_dicts(dict1, dict2):
    merged = {}
    for d in [dict1, dict2]:
        for key, value in d.items():
            if key in merged:
                merged[key].append(value)
            else:
                merged[key] = [value]
    return merged
dict1 = {'a': 1, 'b': 2, 'c': 3}
dict2 = {'a': 4, 'b': 5, 'd': 6}
merged_dict = merge_dicts(dict1, dict2)
print(merged_dict)
```

```python
def flatten_tuple(nested_tuple):
    flat_list = []
    def flat(t):
        for elem in t:
            if isinstance(elem, tuple):
                flat(elem)
            else:
                flat_list.append(elem)
    flat(nested_tuple)
    return tuple(flat_list)
nested = ((1, 2, (3, 4)), 5, (6, 7, (8, 9)))
print(flatten_tuple(nested))
```

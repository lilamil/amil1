```python
mas = [5, 2, 65, 3, 5, 2, 8, 24, 64, 62, 13, 7, 4, 13, 15]
sum = 0
for i in range(0, len(mas), 2):
  if mas[i] % 2 == 0:
    sum += mas[i]
print(mas)
print('Сумма=', sum)
```

```python
from random import randint
N = int(input('Ввести N: '))
mas = []
for i in range (0, N):
  mas.append(randint(0, 10000))
count = 0
flag = False
for i in range(1, N):
    if mas[i] > mas[i - 1]:
        if not flag:
            count += 1
            flag = True
    else:
        flag = False
print(mas)
print(count)
```

```python
from random import randint
N = int(input('Ввести N: '))
mas = []
for i in range (0, N):
  mas.append(randint(0, 10000))
print('Изначальный массив:', mas)
for i in range ((N-1) // 2 * 2, -1, -2):
  mas.pop(i)
print('Новый массив:', mas)
```

```python
from random import randint
mas = []
sum = 0
for i in range(0, 10):
  mas.append(randint(2,5))
  sum += mas[i]
print('Оценки в классе:', mas)
print('Отличных оценок:', mas.count(5))
print('Хороших оценок:', mas.count(4))
print('Удовлетворительных оценок:', mas.count(3))
print('Неудовлетворительных оценок:', mas.count(2))
print('Средняя оценка:', sum/10)
```

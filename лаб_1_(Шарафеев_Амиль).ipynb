```python
n = int(input('n= '))
lastNum = n%10
countOfThree = 0
countOfLast = 0
countOfEven = 0
while n//10 > 0:
  if n % 10 == 3:
    countOfThree +=1
  if n % 10 == lastNum:
    countOfLast +=1
  if (n % 10) % 2 == 0:
    countOfEven +=1
  n //= 10
print('last=', countOfLast, 'three=', countOfThree, 'even=', countOfEven)

```

```python
n = int(input('n ='))
def primeNumbers(n):
  sum=f"{n}="
  k=0
  while n % 2 == 0:
    k+=1
    n//=2
  sum += "2*"*k
  k = 3
  while n > 1:
    while n % k == 0:
      sum +=f"{k}*"
      n//=k
    k+=2
  return sum
print(primeNumbers(n))
```

```python
def form_b_array(A):
    B = [max(A[i], A[i+1]) for i in range(0, len(A) - 1, 2)]
    return B

N = int(input('N='))
A = [0]*N
for i in range(0,N):
  A[i]=int(input(f'A[{i}]='))
B = form_b_array(A)

print("Массив A:", A)
print("Массив B:", B)
```

```python
def factorial(n):
    result = 1
    for i in range(2, n + 1):
        result *= i
    return result
def sumF(n):
    S = 0
    t = 1
    for i in range(1, 2*n, 2):
        S += t * factorial(i)
        t = -t
    return S
n = int(input('n='))
sumF(n)
```

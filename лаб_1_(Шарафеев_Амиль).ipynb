```python
from math import log

x = float(input('x='))

def switch_case(case):
    switcher = {
        1: lambda: -x**2 - 1.1*x + 9,
        2: lambda: log(x + 3) / (x**3 + 9)
    }
    return switcher.get(case, lambda: "Error")()

if x <= -3:
    print(switch_case(1))
else:
    print(switch_case(2))
```

```python
while True:
  a = int(input('Вертикаль белого короля A = ')) # белый король
  b = int(input('Горизонталь белого короля B = '))
  c = int(input('Вертикаль черной фигуры C = ')) # черный конь/ладья
  d = int(input('Горизонталь черной фигуры D = '))
  e = int(input('Вертикаль искомого поля E = ')) # куда прыгаем
  f = int(input('Горизонталь искомого поля F = '))
  if all(1 <= x <= 8 for x in (a, b, c, d, e, f)):
    break
  else:
    print('Одно из чисел не является натуральным меньше девяти!')
def switch_case(case):
  return {
    0: (abs(a-e) <= 1 and abs(b-f) <= 1) and not((abs(c-e) == 2 and abs(d-f) == 1) or (abs(c-e) == 1 and abs(d-f) == 2)), # для коня
    1: (abs(a-e) <= 1 and abs(b-f) <= 1) and not ((c == e and d != f) or (d == f and c != e)) # для ладьи
  }.get(case, "Error")
print("Король имеет право перейти на эту клетку, если на координатах стоит конь = ", switch_case(0))
print("Король имеет право перейти на эту клетку, если на координатах стоит ладья = ", switch_case(1))
```
